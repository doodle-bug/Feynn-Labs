```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("EV_Vehicle_Data.csv")
```

```python
df.head()
```

```python
print(df.columns.tolist())
```

```python
#filling empty data with 0
df=df.fillna('0')
```

```python
df.info()
```

```python
df.corr()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True)
```

```python
sns.countplot(x = 'Vehicle Type', data = df)
```

```python
sns.countplot(x = 'No. of seats', data = df)
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x = 'No. of seats', hue='Vehicle Type', data=df)
```

```python
plt.figure(figsize=(18,10))
sns.countplot(y = 'Make', data = df)
```

```python
sns.relplot(x="Battery Size (kWh)", y="Top Speed (km/hr)", height=6,hue="Vehicle Type",data=df)
```

```python
sns.relplot(x="Battery Size (kWh)", y="Top Speed (km/hr)", size="No. of seats", height=6,sizes=(15, 100),data=df)
```

```python
sns.relplot(x="Top Speed (km/hr)", y="Driving Range (km)",height=6, hue="Vehicle Type",data=df)
```

```python
sns.jointplot(x=df["Battery Size (kWh)"], y=df["Driving Range (km)"], kind="hex", color="#4CB391")
```

```python
sns.relplot(x="Charging Time (hours)", y="Driving Range (km)", height=6,data=df)
```
